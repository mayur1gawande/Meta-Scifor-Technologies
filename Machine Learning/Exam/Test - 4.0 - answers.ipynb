2- Explain the difference between standard deviation and variance. How do they measure data dispersion?

Variance (s**2 or sigma**2) shows the spread of the data in a sample or population. It shows the average value of datapoints away from the mean.
To calculate the variance: var(x) = (x_i - mean(x))**2 /len(x)
Variance is the sum of square of all differences between datapoint and mean divide by the total datapoints.

Standard deviation (s or sigma) is the square root of variance. It shows the spread of datapoints at a interval called as the deviation.
To calculate the std : std(x) = var(x)**(1/2)

## Section C

```python
Data= [25, 28, 30, 29, 27, 45, 26, 31, 120, 32, 30, 29, 28, 27, 26, 25, 90]
```

```python
import pandas as pd

df = pd.Series(Data)
```

```python
q1 = df.quantile(0.25)
print('First Quartile(Q1): ',q1)
```

```python
q3 = df.quantile(0.75)
print('Third Quartile(Q3): ',q3)
```

```python
iqr = q3 - q1
print('Interquartile Range(IQR): ',iqr)
```

```python
lower_bound = q1 - (1.5*iqr)
print('Lower Bound: ',lower_bound)
```

```python
upper_bound = q3 + (1.5*iqr)
print('Upper Bound: ',upper_bound)
```

```python
lower_side_outliers = df[df<lower_bound]
print('Lower Bound outliers:\n',lower_side_outliers)
print('\nThere are no lower bound outliers in the dataset')
```

```python
upper_side_outliers = df[df>upper_bound]
print('Upper Bound Outliers:',upper_side_outliers)
```
